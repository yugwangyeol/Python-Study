### if문
- 조건에 따라서 다른 코드를 실행하고 싶을 때, `if`문을 사용한다.
- `if`문은 기본적으로 조건문과 조건에 다른 실행문으로 이루어져 있다.
- `if`문 안의 문장들은 조건문의 결과가 `True`인 경우에만 실행된다.
- 반드시 들여쓰기와 조건문 끝에 콜론(:)을 지켜주어야 한다.

```python
if 조건문1:
    실행문1
elif 조건문2:
    실행문2
else:
    실행문3
```

````python
if True:
    print('참입니다.')
````

````python
if False:
    print('참입니다.')
````

# 반복문

### while문
- 조건문이 거짓이 될 때까지 실행문 반복
- 조건문의 결과가 항상 참인 경우 무한루프에 빠질 수 있음
- 반드시 종료조건을 명시할 것

```python
while 조건문:
    실행문
```

````python
while False:
    print('참입니다')
````

````python
while True:
    print('참입니다')
````

**[실습1]**<p>
"콜라를 사먹기 위해서 자판기에 투입한 돈이 1200원이 될 때까지 100원짜리 동전을 넣자"

# zip과 enumerate

### zip
- 길이가 같은 여러 iterable 객체를 받아서 같은 index끼리 tuple로 묶어주는 함수
```python
list(zip(iterable1, iterable2))
```

### enumerate
- 순서가 있는 자료형을 받아 자료의 index와 value를 함께 반환하는 함수
```python
list(enumerate(iterable1, iterable2))
```

# Comprehension

- 파이썬에서 보편적으로 사용하는 iterable한 객체를 간단하게 생성하기 위한 방법
- list, set, dictionary의 자료형을 생성 가능
- 사용하면 코드가 컴팩트해지는 장점이 있지만 너무 복잡하면 가독성이 떨어짐

### list comprehension
```python
변수 = [출력표현식 for 요소 in 입력Sequence [if 조건식]]
evens = [x*2 for x in range(11)]
```

### set comprehension
```python
변수 = {출력표현식 for 요소 in 입력Sequence [if 조건식]}
evens = {x * 2 for x in range(11)}
```

### dict comprehension
```python
변수 = {Key:Value for 요소 in 입력Sequence [if 조건식]}
score_tuples = [('math', 90), ('history', 80), ('english', 95), ('computer engineering', 100)]
score_dict = {t[0]: t[1] for t in score_tuples}
```

**[실습1]**<p>
1부터 20까지 들어있는 리스트를 만드시오.

````python
# 기존의 for문
num = []
for i in range(1,21):
    num.append(i)
num
````

````python
# comprehension을 사용
num = [i for i in range(1,21)]
num
````

**[실습2]**<p>
3의 배수를 제외하고 1에서 20까지 자연수가 있는 리스트를 만드시오.

````python
lst = [i for i in range(1,21) if i % 3 != 0]
lst
````
