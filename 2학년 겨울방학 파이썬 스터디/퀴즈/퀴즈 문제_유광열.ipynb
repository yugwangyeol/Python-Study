**Q1.** 조건에 맞게 2차원 리스트의 값을 넣으시오. (조건문 순서 & list 문제)

    [[2의 배수],[3의 배수],[5의 배수],[7의 배수],[동시에 2개 숫자의 배수],[동시에 3개 숫자의 배수],[동시에 4개 숫자의 배수],[동시에 4개 숫자의 배수], [그외 숫자]]

```python
### 주어진 코드
inp = [i for i in range(1,250)]
oup = [[],[],[],[],[],[],[],[]]
```

```python
# 풀이
for i in inp:
    if i%2 == 0 and i%3 == 0 and i%5 == 0 and i%7 == 0:
        oup[6].append(i)
    elif (i%2 == 0 and i%3 == 0 and i%5 == 0) or (i%2 == 0 and i%3 == 0 and i%7 == 0) or (i%2 == 0  and i%5 == 0 and i%7 == 0) or (i%3 == 0 and i%5 == 0 and i%7 == 0):
        oup[5].append(i)
    elif (i%2 == 0 and i%3 == 0) or (i%2 == 0 and i%5 == 0) or (i%2 == 0 and i%7 == 0) or (i%3 == 0 and i%5 == 0) or (i%3 == 0 and i%7 == 0) or (i%5 == 0 and i%7 == 0):
        oup[4].append(i)
    elif i%2 == 0:
        oup[0].append(i)
    elif i%3 == 0:
        oup[1].append(i)
    elif i%5 == 0:
        oup[2].append(i)
    elif i%7 == 0:
        oup[3].append(i)
    else:
        oup[7].append(i)
oup
```

**Q2.** 다음 n차원을 입력받아 n-1차원 함수를 출력하는 n_func 함수를 만드시오. (함수 & fstring 문제)
    
    예시: '1 + 2x^1 + 3x^2 + 4x^3 + 5x^4' (5를 넣었을때)

```python
### 풀이
num = int(input())

def n_func(x):
    coe = [i+1 for i in range(x)]
    modify = ''
    for i in coe:
        if i == 1:
            modify += '1 + '
        elif i == len(coe):
            modify += f'{i}x^{i-1}'
        else:
            modify += f'{i}x^{i-1} + '
    return modify
n_func(num)
```

**Q5.** 데이터 손실이 가장 작은 방법을 생각하여 search_train, session_train, product_train, master를 합치세요. (merge활용; 이거는 이중성이 있어서 고려)

```python
import pandas as pd
```

```python
search_train = pd.read_csv('search_train.csv')
session_train = pd.read_csv('session_train.csv')
product_train = pd.read_csv('product_train.csv')
master = pd.read_csv('master.csv')
```

```python
train = product_train.merge(session_train,on=['CLNT_ID','SESS_ID'],how='left').merge(search_train,on=['CLNT_ID','SESS_ID']).merge(master,on='PD_C',how='left')
display(train.head())
```

**Q6.** 고객아이디 별로 여러 조건의 열이 들어간 새로운 데이터 프레임을 만드시오 (groupby& agg 활용)

    - 고객 아이디별 총 구매액
    - 고객 아이디별 평균 구매액
    - 고객 아이디별 최대 구매액
    - 고객 아이디별 최소 구매액
    - 고객 아이디별 CLAC1_NM 구매 갯수

```python
train.PD_BUY_AM = train.PD_BUY_AM.apply(lambda x: x.replace(',','')).astype(int)
feature = train.groupby('CLNT_ID')['PD_BUY_AM'].agg([('총구매액','sum'),('평균 구매액','mean'),('최대 구매액','max'),('최소 구매액','min')]).reset_index();feature
```

```python
f = train.groupby('CLNT_ID')['CLAC1_NM'].count().reset_index()
feature = pd.merge(feature,f,on='CLNT_ID',how='left');feature
```
