**C1.** 위에 데이터를 가지고 user_id가 NaN인 행은 모두 지우고, platform과 browser는 최빈값으로 대체하시오.\
(결측치를 채운 후에 인덱스 번호를 다시 생성하시오)  (2점)

```python
info_login['user_id'].dropna()
```

```python
info_login = info_login.iloc[info_login['user_id'].dropna().index,:].reset_index()
del info_login['index']
```

```python
most_platform = info_login['platform'].value_counts(dropna =  True).idxmax()
info_login['platform'] = info_login['platform'].fillna(most_platform)
```

```python
most_browser = info_login['browser'].value_counts(dropna =  True).idxmax()
info_login['browser'] = info_login['browser'].fillna(most_browser)
info_login
```
