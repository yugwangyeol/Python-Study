**Q1.** info_login데이터를 다운 받아 모든 열이 NaN인 행을 모두 지운 후에 c_time열을 분을 모두 00으로 바꾸시오

    c_time 예시) 2018-09-23 1:30 -> 2018-09-23 1:00

```python
info_login = pd.read_csv('info_login.csv',encoding='cp949').iloc[:148999,1:]
info_login.c_time = info_login.c_time.apply(lambda x: x.replace(x[-2:],"00"))
info_login
```

**Q2.** 위에 데이터를 가지고 user_id가 NaN인 행은 모두 지우고, platform과 browser는 최빈값으로 대체하시오.\
(결측치를 채운 후에 인덱스 번호를 다시 생성하시오)

```python
info_login = info_login.iloc[info_login['user_id'].dropna().index,:].reset_index()
del info_login['index']
```

```python
most_platform = info_login['platform'].value_counts(dropna =  True).idxmax()
info_login['platform'] = info_login['platform'].fillna(most_platform)
```

```python
most_browser = info_login['browser'].value_counts(dropna =  True).idxmax()
info_login['browser'] = info_login['browser'].fillna(most_browser)
info_login
```

**Q3.** 위에 만든 info_login데이터를 가지고 login_id와 user_id를 int형으로 바꾸세요. (info_login을 출력할때 바뀐 형태로 출력되어야 함)

```python
info_login.login_id = info_login.login_id.astype(int)
info_login.user_id = info_login.user_id.astype(int)
info_login
```

**Q4.** c_time을 가지고 year, month, day열을 생성하시오.

```python
### 날짜

info_login['year'] = info_login.c_time.apply(lambda x: x[:4])
info_login['month'] = info_login.c_time.apply(lambda x: x[5:7])
info_login['day'] = info_login.c_time.apply(lambda x: x[8:10])
info_login
```
