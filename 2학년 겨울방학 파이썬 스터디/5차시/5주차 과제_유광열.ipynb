```python
import pandas as pd
import numpy as np
```

```python
train_profiles = pd.read_csv('train_profiles.csv',encoding='cp949')
train_click = pd.read_csv('train_clickstreams.tab',sep='\t',encoding='cp949')
```

```python
display(train_profiles)
```

```python
display(train_click)
```

**Q1.** train_profiles data에서 성별과 연령대  열을 새로 만들어 추가하시오

```python
train_profiles['sex'] = train_profiles.GROUP.apply(lambda x: x[0])
train_profiles['age'] = train_profiles.GROUP.apply(lambda x: x[1:])
display(train_profiles)
```

**Q2.** train_profiles data에서 나이와 성별에 따른 cus_id를 카운트 한 피벗 테이블을 만드시오

```python
pd.pivot_table(train_profiles,values='CUS_ID',index='sex',columns = 'age',aggfunc = 'count')
```

**Q3.** train_clickstreans data에서 시간 별로 cus_id의 접속 수를 구하시오.

```python
train_click['hour'] = train_click.TIME_ID.astype(str).apply(lambda x: x[8:]).astype(int)
train_click.groupby('hour')['CUS_ID'].count()
```

**Q4.** train_clickstream data에서 cus_id별 site_cnt 와 st_time의 평균을 구하시오

```python
f = train_click.groupby('CUS_ID')['SITE_CNT','ST_TIME'].mean();f
```

**Q5.** train_clickstream data에서 cus_id별 접속 횟수를 구하여 위에 데이터 프레임 붙인 값을 출력하시오

```python
f1 = train_click.groupby('CUS_ID')['TIME_ID'].count()
teature = pd.merge(f,f1,on='CUS_ID',how='left');teature
```
