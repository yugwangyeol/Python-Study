```python
import pandas as pd
import numpy as np
```

**[데이터 명세서]**
- CUS_ID : 사용자 번호
- TIME_ID : 사이트 접속 시간(클릭한 시점)
- SITE : 사용자가 접속한 사이트
- SITE_CNT : 사이트를 클릭한 횟수
- ST_TIME : 사이트 체류시간(초)
- SITE_NM : 사이트 이름
- BACT_NM : 사이트 대분류
- MACT_NM : 사이트 중분류

- GROUP : 사용자의 성별과 연령대

**Q1.** train_profiles와 train_clickstreams를 불러와 합쳐 df라는 데이터를 만드시오. (1점)

```python
# code here
data_1 = pd.read_csv('train_profiles.csv', encoding = 'cp949')
data_2 = pd.read_csv('train_clickstreams.tab', sep = '\t', encoding = 'cp949')
```

```python
df = pd.merge(data_1, data_2, on = 'CUS_ID')
df.head()
```

<font color='blue'>
    정답입니다.

**Q2.** 성별과 연령대 column을 만드시오. (1점)

```python
# code here
df['SEX'] = df['GROUP'].str[:1]
df['AGE'] = df['GROUP'].str[1:]
df.head()
```

<font color='blue'>
    정답입니다.

**Q3.** TIME_ID를 year, month, day,time으로 나눈 4개의 column을 추가하시오. (2점)

```python
# code here
df['TIME_ID'] = pd.to_datetime(df.TIME_ID, format = '%Y%m%d%H')
df['YEAR'] = df['TIME_ID'].dt.year
df['MONTH'] = df['TIME_ID'].dt.month
df['DAY'] = df['TIME_ID'].dt.day
df['TIME'] = df['TIME_ID'].dt.hour
df.head()
```

    정답입니다.

**Q8.** ST_TIME의 사분위수를 기준으로 ST_TIME을 그룹화하고, 각 그룹의 평균 SITE_CNT을 구하시오. (2점)

**Q9.** 계절별 평균 SITE_CNT를 구하시오. 단, 봄은 3월~ 5월, 여름은 6월~ 8월, 가을은 9월~ 11월, 겨울은 12월~2월이다. (2점)

**Q13**. CUS_ID별 (BACT_NM == 뉴스/미디어인 ST_TIME) / (BACT_NM == 쇼핑인 ST_TIME)을 구하시오. (4점)
