```python
import pandas as pd
import numpy as np
```

**[데이터 명세서]**
- CUS_ID : 사용자 번호
- TIME_ID : 사이트 접속 시간(클릭한 시점)
- SITE : 사용자가 접속한 사이트
- SITE_CNT : 사이트를 클릭한 횟수
- ST_TIME : 사이트 체류시간(초)
- SITE_NM : 사이트 이름
- BACT_NM : 사이트 대분류
- MACT_NM : 사이트 중분류
- ACT_NM : 사이트 소분류
- GROUP : 사용자의 성별과 연령대

**Q1.** train_profiles와 train_click을 불러와 합쳐 df라는 데이터를 만드시오. (1점)

```python
train_profiles = pd.read_csv('train_profiles.csv',encoding='cp949')
train_click = pd.read_csv('train_clickstreams.tab',sep='\t',encoding='cp949')
df = pd.merge(train_click, train_profiles, on ='CUS_ID')
df
```

**Q2.** 성별과 연령대 column을 만드시오. (1점)

```python
df['GENDER'] = df['GROUP'].apply(lambda x : x[0])
df['AGE'] = df['GROUP'].apply(lambda x : x[1:])
df.head()
```

**Q3.** TIME_ID를 year, month, day,time으로 나눈 4개의 column을 추가하시오. (2점)

```python
df['TIME_ID'] = pd.to_datetime(df['TIME_ID'], format = '%Y%m%d%H')
df['YEAR'] = df['TIME_ID'].dt.year
df['MONTH'] = df['TIME_ID'].dt.month
df['DAY'] = df['TIME_ID'].dt.day
df['HOUR'] = df['TIME_ID'].dt.hour
df.head()
```

**Q4.** 30대 여성들의 주로 접속한 사이트 중분류는 무엇인가? 상위 5개만 리스트로 나열하시오. (2점)

```python
df.query('GROUP == "F30"')['MACT_NM'].value_counts().index[:5].to_list()
```

**Q5.** 사람들이 가장 많이 방문한 대분류 사이트의 총체류시간은 얼마인가? (2점)

```python
most_bact_nm = df['BACT_NM'].value_counts().index[0]
df.query('BACT_NM == @most_bact_nm')['ST_TIME'].sum()
```

**Q6.** 성별별 평균 사이트 접속 시간을 구하시오. (1점)

```python
df.groupby('GENDER')['ST_TIME'].agg([('ST_TIME_MEAN','mean')]).reset_index()
```

**Q7.** 시간별로 사용자의 접속수를 구한 ACCESS_CNT를 dataframe 형태로 출력하시오. (1점)

```python
df.groupby('HOUR')['CUS_ID'].agg([('ACCESS_CNT','count')]).reset_index()
```

**Q8.** ST_TIME의 사분위수를 기준으로 ST_TIME을 그룹화하고, 각 그룹의 평균 SITE_CNT을 구하시오. (2점)

```python
Q1 = df['ST_TIME'].quantile(0.25)
Q2 = df['ST_TIME'].quantile(0.5)
Q3 = df['ST_TIME'].quantile(0.75)
Q4 = df['ST_TIME'].quantile(1)

df['ST_TIME_group'] = df['ST_TIME'].apply(lambda x : 'Q1' if x < Q1 else ('Q2' if x < Q2 else ('Q3' if x < Q3 else 'Q4')))
df.groupby('ST_TIME_group')[['SITE_CNT']].mean().reset_index()
```

**Q9.** 계절별 평균 SITE_CNT를 구하시오. 단, 봄은 3월~ 5월, 여름은 6월~ 8월, 가을은 9월~ 11월, 겨울은 12월~2월이다. (2점)

```python
df['SEASON'] = df['MONTH'].apply(lambda x : "봄" if 3<=x<=5 else ( "여름" if 6<=x<=8 else ( "가을" if 9<=x<=11 else "겨울" )))
df.groupby('SEASON')['SITE_CNT'].mean().reset_index()
```

**Q10.** 사용자별 사이트 체류시간의 최대, 최소, 평균,  표준편차를 구하시오.(단, 열이름을 새로 지정해야 합니다.) (2점)

```python
df.groupby('CUS_ID')['ST_TIME'].agg([('ST_TIME_MIN','min'), ('ST_TIME_MAX','max'), 
                                     ('ST_TIME_MEAN','mean'), ('ST_TIME_STD','std')]).reset_index()
```

**Q11.** 고객의 월별 사이트 총체류시간을 구하시오. (단, 결측값은 0으로 채우세요.) (2점)

```python
pd.pivot_table(df, values='ST_TIME', index='CUS_ID', columns='MONTH', 
               aggfunc= 'sum' ,fill_value = 0).reset_index()
```

**Q13**. CUS_ID별 (BACT_NM == 뉴스/미디어인 ST_TIME) / (BACT_NM == 쇼핑인 ST_TIME)을 구하시오. (단, 결측값은 평균값으로 채우시오.) (4점)

```python
df_nm = df.query('BACT_NM == "뉴스/미디어"').groupby('CUS_ID')['ST_TIME'].agg([('NM','sum')]).reset_index()
df_sp = df.query('BACT_NM == "쇼핑"').groupby('CUS_ID')['ST_TIME'].agg([('SP','sum')]).reset_index()

f = pd.DataFrame([i for i in range(1,2501)],columns = ['CUS_ID'])
f = f.merge(df_nm, on = 'CUS_ID', how = 'left').merge(df_sp, on = 'CUS_ID', how = 'left')
f = f.fillna(f.mean())
f['NM/SP'] = f['NM']/f['SP'].fillna(0)
f
```

**Q14.** 13번에서 구한 값을 정규화하시오. (구글링) (3점)

```python
x_max = f['NM/SP'].max()
x_min = f['NM/SP'].min()

f['NM/SP_NOR'] = f['NM/SP'].apply(lambda x : (x - x_min)/(x_max - x_min))
f
```
