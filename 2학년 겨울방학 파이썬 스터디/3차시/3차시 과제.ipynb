**Q1.**  회문(palindrome)이란 거꾸로 써도 똑같은 말이 되는 단어를 의미한다.
```python
'다시합창합시다', '수박이박수', '소주만병만주소', 'racecar'
```
입력받은 문자열이 회문인지 아닌지 판단하는 함수 palin()을 작성하시오. 단, 띄어쓰기는 무시한다.

````python
# code here

def palin(string):
    return string==string[::-1]
    
print(palin('토마토'))
    
````

**Q2.** n차원에 있는 두 점 a,b의 거리를 구하는 함수 distance를 만드세요

````python
### 코드를 건드리지 마시고 실행하세요

a = [1,8,-2,4,5]
b = [2,4,-10,7,-5]
````

````python
### 코드를 입력하세요

def distance(a,b):
    dist_1=0
    for i in range(1,len(a)-1):
        dist_1=dist_1+(b[i]-a[i])**2
    return (dist_1)**(1/2)
````

````python
### 결과 확인

distance(a,b)
````

**Q3.** 다음 숫자를 받아 각 자릿수의 합을 구하는 함수를 만드시오.(단 입력된 숫자가 음수일 경우 각 자리수의 합*(-1)을 출력한다.)

    예시1) 입력: 1234   출력: 10
    예시2) 입력: -101    출력: -2

````python
def f():
    number = input('수를 입력하시오:')
    number = ','.join(number)
    a = number.split(',')
    if a[0] == '-':
        a.remove(a[0])
        a = map(int,a)
        ssum = (-1)*sum(a)
    else:
        a = map(int,a)
        ssum = sum(a)
    return ssum
f()
````
