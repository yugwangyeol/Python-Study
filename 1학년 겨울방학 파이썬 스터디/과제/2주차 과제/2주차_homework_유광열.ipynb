<br>trans 데이터는 온라인 쇼핑몰의 transaction을 보여주는 데이터다.</br>
<br>clnt_id는 고객번호를 뜻하며 고객을 구분해준다. pd_c는 상품 코드를 뜻하고 상품을 구분해준다.</br>
<br>clnt_id의 고유한 값의 개수는 11271개지만 값의 범위는 2 ~ 72424 이다.</br>
<br>pd_c의 고유한 값의 개수는 1667개지만 값의 범위는  1 ~ 1668 이다.</br>
<br>특정 패키지를 사용할 때 이 값들의 범위를 0 ~ (고유값의 개수 - 1)로 만들어 줘야 하는 경우가 생긴다.</br>
<br>이렇게 데이터를 다른 형식으로 만드는 과정을 인코딩이라고 한다.</br>

<br>[문제] trans 데이터 프레임의 clnt_id 와 pd_c를 인코딩 해서 새로운 데이터 프레임을 생성하시오</br>
<br>기존 데이터프레임과 새로 만들어진 데이터프레임의 shape는 같아야 한다.</br>

```
clnt_id = 39423 -> clnt_id = 0, pd_c = 0565 -> pd_c = 0
clnt_id = 21279 -> clnt_id = 1, pd_c = 0565 -> pd_c = 0
clnt_id = 48969 -> clnt_id = 2, pd_c = 0572 -> pd_c = 1
clnt_id = 30533 -> clnt_id = 3, pd_c = 0670 -> pd_c = 2
clnt_id = 64346 -> clnt_id = 4, pd_c = 0543 -> pd_c = 3
```
```
                                        
                                                            변경 전
```
| |clnt_id|pd_c|

```
                                                        clnt_id: 2 ~ 72424

                                                            변경 후
```

````python
## code here

cln=pd.DataFrame(list(set(trans['clnt_id'])))
cln.columns = ['clnt_id']
cln['c']=range(0,len(cln))
cln.head()

pdc=pd.DataFrame(list(set(trans['pd_c'])))
pdc.columns = ['pd_c']
pdc['d']=range(0,len(pdc))
pdc.head()

trans_1=pd.merge(trans,cln, on='clnt_id',how='left')
trans_2=pd.merge(trans_1,pdc, on='pd_c',how='left');trans_2
trans_2.drop(['clnt_id','pd_c'],axis=1,inplace=True)
trans_2.rename(columns={'c':'clnt_id','d':'pd_c'},inplace=True);trans_2
trans_2.shape

# 코드가 너무 더러운 답이랍니다 ㅠㅠ
````
