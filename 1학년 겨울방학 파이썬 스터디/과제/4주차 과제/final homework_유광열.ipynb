```python
import pandas as pd
import numpy as np
```

```python
dt1 = pd.read_table('train_clickstreams.tab', encoding = 'cp949')
dt2 = pd.read_csv('train_profiles.csv', encoding = 'cp949')
```

#### 1. dt1과 dt2를 합치시오.

```python
dt3=pd.merge(dt1,dt2,on='CUS_ID',how='left');dt3
```

#### 2. 데이터의 GROUP열을 성별 column, 나이 column으로 나누고, GROUP 열을 삭제하시오.

```python
dt3['성별']=dt3.GROUP.str.get(i=0)
dt3['나이']=dt3.GROUP.str.slice(start=1,stop=5)
dt3.drop('GROUP',axis=1,inplace=True);dt3
```

#### 3. TIME_ID column을 연, 월, 일, 시 column으로 나누고, 요일 정보를 담고 있는 weekday column을 만든 후, TIME_ID column을 제거하시오.

```python
dt3.TIME_ID=pd.to_datetime(dt3.TIME_ID,format='%Y%m%d%H')
dt3['년']= dt3['TIME_ID'].dt.strftime('%y');dt3
dt3['월']= dt3['TIME_ID'].dt.strftime('%m');dt3
dt3['일']= dt3['TIME_ID'].dt.strftime('%d');dt3
dt3['시']= dt3['TIME_ID'].dt.strftime('%H');dt3
dt3['weekday']=dt3['TIME_ID'].dt.weekday_name
```

```python
dt3.drop('TIME_ID',axis=1,inplace=True);dt3
```

#### 4. CUS_ID별 SITE_CNT 횟수의 총합을 구하시오.

```python
dt3.groupby('CUS_ID')['SITE_CNT'].sum()
```

#### 5. 남자, 여자의 평균 SITE_CNT 횟수를 구하시오.

```python
dt3.groupby('성별')['SITE_CNT'].mean()
```

#### 6. 연령대별 평균 ST_TIME을 구하시오.

```python
dt3.groupby('나이')['ST_TIME'].mean()
```

#### 7. 문제 3에서 만든 시간 column을 이용해, 시간대별 평균 SITE_CNT를 구하시오. 단, 시간대는 0시~5시 / 6시 ~ 11시 / 12시 ~ 17시 / 18시 ~ 23시로 구분된다.

```python
dt3['시']=pd.to_numeric(dt3['시'])

a=dt3.query('시>=0 & 시<=5').SITE_CNT.mean()
b=dt3.query('시>=6 & 시<=11').SITE_CNT.mean()
c=dt3.query('시>=12 & 시<=17').SITE_CNT.mean()
d=dt3.query('시>=18 & 시<=23').SITE_CNT.mean()
print(a,b,c,d)
```

#### 8. CUS_ID의 BACT_NM별 ST_TIME의 총합을 구하시오.

```python
BS=pd.DataFrame(dt3.groupby(['CUS_ID','BACT_NM'])['ST_TIME'].sum());BS
```

```python
CBS=pd.pivot_table(dt3,values='ST_TIME',index='CUS_ID',columns='BACT_NM',aggfunc=sum);CBS
```

#### 9. CUS_ID별 (BACT_NM == 뉴스/미디어인 ST_TIME) / (BACT_NM == 쇼핑인 ST_TIME)을 구하시오.

```python
n_s=pd.DataFrame(CBS.loc[:,('뉴스/미디어')]/CBS.loc[:,('쇼핑')])
n_s
```

#### 10. 9에서 구한 값을 정규화시키시오.

```python
n_s.fillna(n_s.mean(),inplace=True)
n_s.isnull().sum()
np.nan_to_num(n_s,copy=False)

from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
fitted = min_max_scaler.fit(n_s)
print(fitted.data_max_)

n_s = min_max_scaler.transform(n_s)
n_s = pd.DataFrame(n_s)
print(n_s.head())
```
