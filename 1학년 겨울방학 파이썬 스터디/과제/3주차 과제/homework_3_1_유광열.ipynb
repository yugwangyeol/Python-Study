Q4. `축산물`은 하루 중 언제 가장 많이 팔리는가?

```python
data_buy_1=data_buy.sort_values(by='구매수량',ascending=False)
data_buy_1.query('상품중분류명=="축산물"',inplace=True)
data_buy_1.iloc[0,5]
```

Q5. `총 구매액`이 가장 많은 사람부터 적은 사람 순으로 정렬하여 `상위 10명`만 출력하시오. 

```python
buy=pd.DataFrame(data_buy.groupby('ID')['구매금액'].sum())
buy_1=pd.merge(data_buy,buy,on='ID',how='left');buy_1
buy_1.rename(columns={'구매금액_x':'구맥금액','구매금액_y':'총구매액'},inplace=True)
buy_1.sort_values(by='총구매액', ascending=False).head(10)
```

Q6. `고객`의 `구매지역`별 `구매금액`의 합을 구하시오.
