## Duplicate Rows

```python
data = pd.DataFrame({
    'key1':['a', 'b', 'b', 'c', 'c'],
    'key2':['v', 'w', 'w', 'x', 'y'],
    'num':[1, 2, 2, 4, 5]
})
data
```

```python
## 중복여부 확인
data.duplicated()
```

```python
data.duplicated(['key1'])
```

```python
data.duplicated(['key1','key2'])
```

```python
## 중복값 처리
data.drop_duplicates()
```

```python
data.drop_duplicates(['key1']) # keep='first' default
```

```python
data.drop_duplicates(['key1','key2'])
```

```python
## 중복된 마지막 값을 남김
data.drop_duplicates(['key1'],keep='last')
```

## Merging DataFrame

<font color='darkgreen'>
##### Merge Types

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
```

```python
display(df1,df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1, df2, on='key', how='outer')
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, on='key', how='right')
```

- column명이 다른 경우

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
```

```python
display(df3,df4)
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

## Concatenating DataFrame

```python
df5 = pd.DataFrame(np.arange(6).reshape(3,2),
                  index=['a','b','c'], columns=['one', 'two'])
df6 = pd.DataFrame(5+np.arange(4).reshape(2,2),
                  index=['a','c'], columns=['three','four'])
```

```python
display(df5, df6)
```

```python
pd.concat([df5,df6], axis=1) ## 열로 추가
```

```python
df7 = pd.DataFrame(np.arange(12).reshape(3,4), columns=['a', 'b', 'c', 'd'])
df8 = pd.DataFrame(12+np.arange(6).reshape(2,3), columns=['b','d','a'])
```

```python
display(df7, df8)
```

```python
pd.concat([df7,df8]) ## 행으로 추가
```

```python
pd.concat([df7, df8], ignore_index=True) ## 겹치는 index 무시
```

## Handling Missing Data

```python
data = pd.merge(df3, df4, left_on='lkey', right_on='rkey',how='outer')
data
```

- column별 / row별 결측값 개수 구하기

```python
data.isnull()
# pd.isnull(data)
```

```python
## column별 개수
data.isnull().sum()
# pd.isnull(data).sum()
```

```python
## row별 개수
data.isnull().sum(1)
# pd.isnull(data).sum(1)
```

```python
data.notnull()
# pd.notnull(data)
```

```python
## column별 개수
data.notnull().sum()
# pd.notnull(data).sum()
```

```python
## row별 개수
data.notnull().sum(1)
# pd.notnull(data).sum(1)
```

- 결측값 채우기

```python
data.fillna(0)
```

```python
data.fillna('missing')
```

```python
## 결측값을 앞 방향으로 채우기
data.fillna(method='ffill')
# data.fillna(method='pad')
```

```python
## 결측값을 뒷 방향으로 채우기
data.fillna(method='bfill')
```

```python
## column별 평균으로 대체
data.fillna(data.mean())
```

```python
## column마다 다른 값으로 대체
data.fillna({'lkey':'ㅋ','data1':100,'rkey':'ㅋㅋ','data2':-100})
```

- 결측값이 들어있는 row / column 제거하기

```python
data.dropna(axis=1)
```

```python
data.dropna(how='all')
```

```python
data.dropna(how='all', axis=1)
```
