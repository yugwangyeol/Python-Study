```python
import pandas as pd
import numpy as np
```

```python
data_id = pd.read_csv('Id_info.csv', encoding = 'cp949')
data_buy = pd.read_csv('Buy_info.csv', encoding = 'cp949')
data = pd.merge(data_id, data_buy)
```

Q1. `70대 여성` 고객정보를 출력하시오.

```python
data_id.query('성별 == 2 and 연령 >= 70 and 연령 < 80')
```

Q2. `남자` 고객과 `여성` 고객의 평균 나이는?

```python
data_id.groupby('성별')['연령'].mean()
```

Q3. `상품중분류명` 기준 상품종류의 개수는?

```python
data['상품중분류명'].nunique()
```

Q4. `축산물`은 하루 중 언제 가장 많이 팔리는가?

```python
data.query('상품중분류명 == "축산물"')['구매시간'].value_counts()[:1]
```

```python
data.query('상품중분류명 == "축산물"').groupby('구매시간')['구매수량'].sum().sort_values(ascending = False)[:1]
```

```python
data.query('상품중분류명 == "축산물"').groupby('구매시간')['구매금액'].sum().sort_values(ascending = False)[:1]
```

Q5. `총 구매액`이 가장 많은 사람부터 적은 사람 순으로 정렬하여 `상위 10명`만 출력하시오. 

```python
data.groupby('ID')[['구매금액']].sum().sort_values(by='구매금액', ascending=False)[:10]
```

Q6. `고객`의 `구매지역`별 `구매금액`의 합을 구하시오.

```python
pd.pivot_table(data, index = 'ID', columns = '구매지역', values = '구매금액', 
               aggfunc = sum, fill_value = 0)
```
