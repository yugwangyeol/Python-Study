### 새로운 column 생성

```python
# 계절 column 생성
df['구매월'] = df['구매일자'].dt.month
def g(x) : 
    if 3<=x <=5:
        return "Spring"
    elif 6<=x <=8 :
        return "Summer"
    elif 9<=x <=11 :
        return "Fall"
    else :
        return "Winter"
df['계절'] = df['구매월'].apply(g)

df["구매일자"] = pd.to_datetime(df['구매일자'],format='%Y%m%d')
df['개당구매금액'] = (df['구매금액'] / df['구매수량']).astype(int)
```

```python
# 데이터 통합
# 상품중분류명이 농산물, 수산물, 축산물인 데이터들을 식자재로 통합하기
df.loc[df['상품중분류명'] == '농산물', '상품중분류명'] = '식자재'
df.loc[df['상품중분류명'] == '수산물', '상품중분류명'] = '식자재'
df.loc[df['상품중분류명'] == '축산물', '상품중분류명'] = '식자재'
```

# 구매횟수비중, 구매금액비중 확인 Tool

```python
# 전체 DataFrame
s = pd.DataFrame()
s['평균구매금액']=df.groupby('상품중분류명')['구매금액'].mean().round()#.reset_index()
s['구매횟수'] = df.상품중분류명.value_counts()
s = s.reset_index()
t = s['평균구매금액'].sum()
u = s['구매횟수'].sum()
s['구매금액비중'] = s['평균구매금액'] / t * 100
s['구매횟수비중'] = s['구매횟수'] / u * 100
#s
v = s.sort_values(by = '구매횟수비중', ascending = False)
```

```python
f1 = plt.figure(figsize=(20,10))
plt.scatter(v['구매횟수비중'],v['구매금액비중'])
plt.grid(True)
plt.xlabel('구매비중', size = 20)
plt.ylabel('판매금액비중', size = 20)

for i,j,k in zip(v['구매횟수비중'], v['구매금액비중'], v['상품중분류명']):
    plt.text(i,j,k, fontsize = 12)
#plt.plot(np.arange(0,14,0.1), np.arange(0,14,0.1))
plt.show()
```

```python
# 데이터 초기화
demo = pd.read_csv("Demo.csv", encoding = "cp949")
buy = pd.read_csv("Buy_info.csv", encoding = "cp949")
df = pd.merge(demo,buy)
df["구매일자"] = pd.to_datetime(df['구매일자'],format='%Y%m%d')
df['개당구매금액'] = (df['구매금액'] / df['구매수량']).astype(int)
# 이상치 제거
df['구매월'] = df['구매일자'].dt.month
def g(x) : 
    if 3<=x <=5:
        return "Spring"
    elif 6<=x <=8 :
        return "Summer"
    elif 9<=x <=11 :
        return "Fall"
    else :
        return "Winter"
df['계절'] = df['구매월'].apply(g)
# 골프용품/의류 이상치 제거
df.drop([39191,39218,39244,39285], inplace= True)

# 명품 이상치 제거
df.drop([35536,45826], inplace= True)

# 화장품 이상치 제거
df.drop([4887,21778], inplace= True)

# 가공식품 이상치 제거
df.drop(98929, inplace= True)

# 건강식품 이상치 제거
df.drop(76739, inplace= True)

# 농산물 이상치 제거
df.drop(46634, inplace= True)

# 수산물 이상치 제거
df.drop([38362,46536], inplace= True)

# 주얼리 이상치 제거
df.drop(72979, inplace= True)

# 구매금액이 0인 행 제거
df = df.query('구매금액 != 0')

# 개당구매금액이 10으로 나누어 떨어지지 않는 데이터 제거
df = df.query('개당구매금액 % 10 == 0')
# 데이터 통합
# 상품중분류명이 농산물, 수산물, 축산물인 데이터들을 식자재로 통합하기
df.loc[df['상품중분류명'] == '농산물', '상품중분류명'] = '식자재'
df.loc[df['상품중분류명'] == '수산물', '상품중분류명'] = '식자재'
df.loc[df['상품중분류명'] == '축산물', '상품중분류명'] = '식자재'
df_bb = df.query('연령 <= 59')
df_nbb = df.query('연령 > 59')
```
