#### 1. dt1과 dt2를 합치시오.

```python
data = pd.merge(dt1, dt2) ; data
```

#### 2. 데이터의 GROUP열을 성별 column, 나이 column으로 나누고, GROUP 열을 삭제하시오.

```python
data['GENDER'] = data['GROUP'].apply(lambda x:x[:1])
data['AGE'] = data['GROUP'].apply(lambda x:x[1:3])
del data['GROUP']
data.head()
```

#### 3. TIME_ID column을 연, 월, 일, 시 column으로 나누고, 요일 정보를 담고 있는 weekday column을 만든 후, TIME_ID column을 제거하시오.

```python
data['TIME_DATE'] = data['TIME_ID'].apply(lambda x: str(x)[:8])
data['TIME_HOUR'] = data['TIME_ID'].apply(lambda x: str(x)[8:])
data['TIME_DATE'] = pd.to_datetime(data['TIME_DATE'])

data['year'] = data['TIME_DATE'].dt.year
data['MONTH'] = data['TIME_DATE'].dt.month
data['DAY'] = data['TIME_DATE'].dt.day
data['HOUR'] = pd.to_datetime(data['TIME_HOUR'], format ='%H').dt.hour
data['weekday'] = data['TIME_DATE'].dt.dayofweek 

del data['TIME_ID'], data['TIME_DATE'], data['TIME_HOUR']
data.head()
```

#### 4. CUS_ID별 SITE_CNT 횟수의 총합을 구하시오.

```python
data.groupby('CUS_ID')['SITE_CNT'].sum()
```

#### 5. 남자, 여자의 평균 SITE_CNT 횟수를 구하시오.

```python
data.groupby('GENDER')['SITE_CNT'].mean()
```

#### 6. 연령대별 평균 ST_TIME을 구하시오.

```python
data.groupby('AGE')['ST_TIME'].mean()
```

#### 7. 문제 3에서 만든 시간 column을 이용해, 시간대별 평균 SITE_CNT를 구하시오. 단, 시간대는 0시~5시 / 6시 ~ 11시 / 12시 ~ 17시 / 18시 ~ 23시로 구분된다.

```python
def time(x):
    if x >=0 and x <=5:
        return 5
    elif x >=6 and x <= 11:
        return 11
    elif x >= 12 and x <= 17:
        return 17
    elif x >= 18 and x <= 23:
        return 23
    else:
        return None
    
data['TM'] = data['HOUR'].apply(lambda x: time(int(x)))
data.groupby('TM')['SITE_CNT'].mean()
```

#### 8. CUS_ID의 BACT_NM별 ST_TIME의 총합을 구하시오.

```python
data1 = pd.pivot_table(data, index = 'CUS_ID', columns = 'BACT_NM', values = 'ST_TIME',
               aggfunc = sum, fill_value = 0)
data1
```

#### 9. CUS_ID별 (BACT_NM == 뉴스/미디어인 ST_TIME) / (BACT_NM == 쇼핑인 ST_TIME)을 구하시오.

```python
# column indexing을 위한 column name 확인
data1.columns

data2 = data1[['뉴스/미디어','쇼핑']]
data2['뉴스/쇼핑'] = data2['뉴스/미디어'] / data2['쇼핑']

# divide by 0 처리
## inf를 제외한 최댓값
MAX = data2.loc[data2['뉴스/쇼핑'] != np.inf, '뉴스/쇼핑'].max()

data2['뉴스/쇼핑'].replace(np.inf, MAX, inplace=True)

# 결측값 처리
data2.fillna(0, inplace = True)

data2
```

#### 10. 9에서 구한 값을 정규화시키시오.

```python
def normalize(x, min_x, max_x): 
    return (x-min_x)/(max_x-min_x)

min_x = data2['뉴스/쇼핑'].min(); max_x = data2['뉴스/쇼핑'].max()

data2['정규화'] = data2['뉴스/쇼핑'].apply(normalize, args=(min_x, max_x))        
```
