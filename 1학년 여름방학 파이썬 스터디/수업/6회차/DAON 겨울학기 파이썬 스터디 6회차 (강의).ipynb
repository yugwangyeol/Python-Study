# 1. Merge 
- 두 개의 데이터프레임 합치기

```python
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})
df3
```

```python
df4 = DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
df4
```

### pd.merge(합칠 데이터프레임, 합칠 키값, 합칠 방식)

```python
df5 = pd.merge(df3,df4,left_on='lkey',right_on='rkey',how='outer')
df5
```

# 3. Apply
- 데이터 프레임에 조건 적용 값 변경하기
- 함수생성후 함수 적용하기

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
        'year': [2000, 2001, 2002, 2001, 2002],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9]}
df = DataFrame(data)
df
```

```python
if df['pop'] > 2:                
    df['pop2'] = df['pop'] + 2
else:
    df['pop2'] = df['pop'] + 1.5
```

```python
def f(x) : 
    if x > 2:
        return x + 2
    else:
        return x + 1.5

df['pop2'] = df['pop'].apply(f)  #apply 권장!!
df
```

```python
def f(x) : 
    if x < 2:
        return "1점대"
    elif  x < 3 :
        return "2점대"
    else :
        return " 3점대"

df['pop3'] = df['pop'].apply(f)  #apply 권장!!
df
```
