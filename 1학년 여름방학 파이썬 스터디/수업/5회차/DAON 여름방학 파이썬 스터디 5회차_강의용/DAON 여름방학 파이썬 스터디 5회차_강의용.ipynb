```python
df.loc[df["year"] > 2000, :]
```

```python
df.loc[df['continent']=="Asia",:]
```

```python
# 아시아 대륙의 lifeExp의 평균
df.loc[df['continent']=='Asia','lifeExp'].sum()/len(df.loc[df['continent']=='Asia','lifeExp'])
```

```python
round(df.loc[df['continent']=='Asia','lifeExp'].mean(axis=0),2)
```

```python
# 아시아 대륙의 lifeExp와 gdpPercap을 뽑아라. 
df.loc[df['continent']=='Asia',['lifeExp','gdpPercap']]
```

## 조건이 여러개일 경우
- & 연산자를이용하여, 여러개의 조건 행 인덱스를 넣어주면된다.

```python
df.loc[(조건1) & (조건2),:]
```

```python
df.loc[(df['year']<2000) & (df['year'] > 1990),:]
```

```python
# year가 2000년 이후면서 lifeExp가 50 이상인 곳을 추출하라. 
df.loc[(df['year']>2000) & (df['lifeExp']>50),:]
```
